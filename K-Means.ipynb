```python
# keeping only the desired column
col = ['Genre', 'Age', 'AnnualIncome', 'SpendingScore']
customer_features = df[col]
```

```python
customer_features = pd.get_dummies(customer_features, 
                                     columns= ['Genre'])
```

```python
customer_features['cluster']=0
```

```python
mean = np.mean(customer_features, axis=0)
mean
```

```python
# Function to initialise centroids

def init(K):
    centroids = []
    r = [randint(0, len(customer_features)) for i in range(0, K)]
    for i in range(K):
        centroids.append(customer_features.iloc[r[i],:])
    return centroids
```

```python
def euclideanDistance(tst,trn):
    d = np.sqrt( (trn[0] - tst[0])**2 + (trn[1] - tst[1])**2 + (trn[2] - tst[2])**2 + 
                 (trn[3] - tst[3])**2 + (trn[4] - tst[4])**2 )
    return d
```

```python
def k_means(centroids,K):
    for i in range(len(customer_features)):
        dist = []
        pt = customer_features.iloc[i,:]
        for j in range(K):
            dist.append(euclideanDistance(pt,centroids[j]))            
        min_val = dist.index(min(dist))
        customer_features['cluster'][i] = min_val + 1
```

```python
# Reassign centroids

def reassign(K):            
    centroids = []
    for i in range(K):
        centroids.append(np.mean(customer_features.loc[ customer_features['cluster'] ==  i+1 ], axis=0))
    return centroids
```

```python
# Stopping criteria

def compare(K):
    c = []
    ctr = 0
    ctr2 = 0
    for i in range(K):
        c.append(np.mean(customer_features.loc[ customer_features['cluster'] ==  i+1 ], axis=0))
#         print(c[i][0])
#         print(centroids[i][0])
#         break
    for i in range(K):
        for j in range(4):
            if c[i][j] == centroids[i][j]:
                ctr += 1
        if ctr == 4:
            ctr2 += 1
            
    if ctr2 == K:
        return False
    
    return True
```

```python
def implement():
    K = 3
    centroids = init(K)    
    c = True
    
    while c == True:
        k_means(centroids,K)
        print(customer_features)
        c = compare(K)
        reassign(K)
```
